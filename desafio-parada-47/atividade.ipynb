Gr√°fico de barras

```python
import matplotlib.pyplot as plt
import pandas as pd

# Leitura do arquivo Excel
tabela = pd.read_excel("tabela_alunos.xlsx")

# Contagem de atividades por data e hora
atividades = tabela.groupby(['Data', 'Hora', 'Atividade']).size().unstack(fill_value=0)

# Plotar o gr√°fico de barras empilhadas
ax = atividades.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.xlabel('Data e Hora')
plt.ylabel('N√∫mero de Pessoas')
plt.title('Atividade de Entrada/Sa√≠da por Data e Hora')
plt.legend(title='Atividade', loc='upper right')
plt.xticks(rotation=45)


plt.tight_layout()

# Exibir o gr√°fico
plt.show()
```
